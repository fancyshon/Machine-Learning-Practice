````python
# Read test set
test_set = pd.read_csv('test.csv')
# 資料分析與前處理

train_x = df[['Name', 'Sex', 'Age','Fare' ,'Pclass', 'Embarked' ,'Parch' ,'SibSp']]              # 取出訓練資料需要分析的資料欄位
train_y = df['Survived']                       # 取出訓練資料的答案
test_x = test_set[['Name', 'Sex', 'Age', 'Fare','Pclass', 'Embarked' ,'Parch' ,'SibSp']]

from sklearn.impute import SimpleImputer       # 匯入填補缺失值的工具
from sklearn.preprocessing import LabelEncoder # 匯入 Label Encoder
from sklearn.preprocessing import OneHotEncoder# One hot encoder

imputer = SimpleImputer(strategy='median')     # 創造 imputer 並設定填補策略
age = train_x['Age'].to_numpy().reshape(-1, 1)
imputer.fit(age)                               # 根據資料學習需要填補的值
train_x['Age'] = imputer.transform(age)        # 填補缺失值
train_x['Age'] = [0 if i <= 16 else 1 if i > 16 and i < 32 else 2 if i > 32 and i < 48 else 3 if i >= 48 and i < 64 else 4 for i in train_x['Age']]

train_x['Fare'] = [0 if i <= 7.91 else 1 if i > 7.91 and i < 14.454 else 2 if i > 14.454 and i <= 31 else 3 if i > 31 and i <= 513 else 4 for i in train_x['Fare']]

# get_dummies 只能用字串
# for i in train_x["Pclass"]:
#     i=str(i)
# train_x = pd.get_dummies(train_x, columns=['Pclass'])

# 填補Embarked
imputer = SimpleImputer(strategy='most_frequent')
embarked = train_x['Embarked'].to_numpy().reshape(-1,1)
imputer.fit(embarked)
train_x['Embarked'] = imputer.transform(embarked)
train_x = pd.get_dummies(train_x, columns=['Embarked'])

le = LabelEncoder()                            # 創造 Label Encoder
le.fit(train_x['Sex'])                         # 給予每個類別一個數值
train_x['Sex'] = le.transform(train_x['Sex'])  # 轉換所有類別成為數值

# 將Name轉換成稱謂
name = train_x['Name']
train_x['Title'] = [i.split(".")[0].split(",")[-1].strip() for i in name]
train_x["Title"] = train_x["Title"].replace(["Dr",'Don','Rev','Lady', 'Sir', 'Col', 'Capt', 'the Countess',             'Jonkheer', 'Dona'], "Other")
train_x["Title"] = ["Master" if i == "Master" else "Mrs" if i == "Mrs" or i == "Mme" else "Mr" if i == "Mr" else "Miss" if i == "Miss" or i == "Ms" or i == 'Mlle' else "Other" for i in train_x["Title"]]
train_x = pd.get_dummies(train_x, columns=['Title'])
train_x.drop(labels = ["Name"], axis = 1, inplace = True)

# 找出這位乘客是否只有自己一位坐船，沒有親戚
train_x['Family_size'] = train_x['SibSp'] + train_x['Parch'] + 1
train_x["Alone"] = [1 if i == 1 else 0 for i in train_x['Family_size']]
train_x = train_x.drop(['Parch', 'SibSp', 'Family_size'], axis=1)


# For test
age = test_x['Age'].to_numpy().reshape(-1, 1)
imputer.fit(age)                              
test_x['Age'] = imputer.transform(age)
test_x['Age'] = [0 if i <= 16 else 1 if i > 16 and i < 32 else 2 if i > 32 and i < 48 else 3 if i >= 48 and i < 64 else 4 for i in test_x['Age']]

test_x['Fare'] = [0 if i <= 7.91 else 1 if i > 7.91 and i < 14.454 else 2 if i > 14.454 and i <= 31 else 3 if i > 31 and i <= 513 else 4 for i in test_x['Fare']]

le = LabelEncoder()                            
le.fit(test_x['Sex'])                         
test_x['Sex'] = le.transform(test_x['Sex'])  

# for i in test_x["Pclass"]:
#     i=str(i)
# test_x = pd.get_dummies(test_x, columns=['Pclass'])

test_x = pd.get_dummies(test_x, columns=['Embarked'])

name = test_x['Name']
test_x['Title'] = [i.split(".")[0].split(",")[-1].strip() for i in name]
test_x["Title"] = test_x["Title"].replace(["Dr",'Don','Rev','Lady', 'Sir', 'Col', 'Capt', 'the Countess',             'Jonkheer', 'Dona'], "Other")
test_x["Title"] = ["Master" if i == "Master" else "Mrs" if i == "Mrs" or i == "Mme" else "Mr" if i == "Mr" else "Miss" if i == "Miss" or i == "Ms" or i == 'Mlle' else "Other" for i in test_x["Title"]]
test_x = pd.get_dummies(test_x, columns=['Title'])
test_x.drop(labels = ["Name"], axis = 1, inplace = True)

test_x['Family_size'] = test_x['SibSp'] + test_x['Parch'] + 1
test_x["Alone"] = [1 if i == 1 else 0 for i in test_x['Family_size']]
test_x = test_x.drop(['Parch', 'SibSp', 'Family_size'], axis=1)

# train_x.head(10)
# test_x.head(10)

````

## Decision tree

- 決策樹（decision tree）是一個樹結構（可以是二叉樹或非二叉樹）。其每個非葉節點表示一個特徵屬性上的測試，每個分支代表這個特徵屬性在某種條件下的結果
- 樹包含三個節點
    - 根節點:沒有任何進入的邊，而且有0個邊或是輸出的邊
    - 內部節點:每個節點都有一個輸入的邊，以及二個或多個輸出的邊
    - 葉節點:每個節點都有一個輸入的邊，但沒有輸出的邊
        - 每個葉節點都是一個**類別標記**


````python
# 模型訓練

from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.tree import DecisionTreeClassifier       # 匯入決策樹模型
from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度


for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    model = DecisionTreeClassifier(random_state=1012 ,splitter='random') # 創造決策樹模型
    model.fit(train_x_split, train_y_split)           # 訓練決策樹模型
    
    train_pred_y = model.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = model.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

final_model = DecisionTreeClassifier(random_state=1012)
final_model.fit(train_x, train_y)
final_pred = final_model.predict(train_x)
final_acc = accuracy_score(train_y,final_pred)


print("Decision tree :")
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}\n').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
print('使用所有資料訓練的準確度: {}'.format(final_acc))
````

## Ensemble learn 

利用多個基本分類器、弱分類器對一的data set去做分類，再將所有分類器產生的結果綜合考慮，得到更好的預測結果。

- Bagging 
    - Bootstrap Aggregating 的簡稱，透過統計學的 Bootstrap sampling 得到不同的訓練資料，然後根據這些訓練資料得到一系列的基本分類器，假如演算法產生了 5 個基本分類器，她們對某個觀測值的預測結果分別為 1, 0, 1, 1, 1，那麼 Bagging 演算法的輸出結果就會是 1，這個過程稱之為基本分類器的投票。
    - Bagging 演算法中大多利用決策樹當作基本分類器

````python
# Ensemble model
from sklearn.ensemble import BaggingClassifier

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度


for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 
    
    bag = BaggingClassifier(n_estimators = 100)       # 產生Bagging模型  
    bag.fit(train_x_split,train_y_split)              # 訓練         
    
    train_pred_y = bag.predict(train_x_split)         # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = bag.predict(valid_x_split)         # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print("Ensemble :")
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}\n').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))

test_bag = BaggingClassifier(n_estimators = 100)
test_bag.fit(train_x,train_y)
ans = bag.predict(test_x)
result = pd.DataFrame({
    'PassengerId' : test_set['PassengerId'],
    'Survived' : ans
})
result.to_csv('submission.csv', index=False)

print('使用所有資料訓練的準確度: {}'.format(accuracy_score(train_y,test_bag.predict(train_x))))
````
